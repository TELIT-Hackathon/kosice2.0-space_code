```python
import numpy as np
import pandas as pd
import geopandas
import os
dataset_path = "_data"
# import pyogrio
```

```python
print("DATASET LOADING -> Schools")
filename = "open_ke_materske.geojson"
file = open(os.path.join(dataset_path,filename))
df_open_ke_materske = geopandas.read_file(file)
filename = "open_ke_stredne.geojson"
file = open(os.path.join(dataset_path,filename))
df_open_ke_stredne = geopandas.read_file(file)
filename = "open_ke_zakladne.geojson"
file = open(os.path.join(dataset_path,filename))
df_open_ke_zaklade = geopandas.read_file(file)
```

```python
print("DATASET MERGE -> Schools")
df_open_ke_materske["type"] = "elementary"
df_open_ke_zaklade["type"] = "primary"
df_open_ke_stredne["type"] = "high"
df_schools = pd.concat([df_open_ke_materske,df_open_ke_zaklade,df_open_ke_stredne])
import random as rand
import string
# df_schools["space_code_id"] = df_schools.apply(id_generator, axis = 1)
df_schools["space_code_id"] = df_schools.apply(lambda row : id_generator(), axis = 1)

def id_generator(size=16, chars=string.ascii_uppercase + string.digits):
    return ''.join(rand.SystemRandom().choice(chars) for _ in range(size))

df_schools.to_csv("schools.csv")
```

```python
print("DATASET LOAD -> nehnutelnosti")
filename = "flats.csv"
file = os.path.join(dataset_path,"nehnutelnosti",filename)
df_rent = pd.read_csv(file)
```

```python
# LOAD BYTY KE
print("DATASET LOAD -> budovy v KE byty, chyze")
df_byty = pd.read_csv(os.path.join(dataset_path, "adresy_pocty_bytov_hack.csv"))
df_byty.drop_duplicates(subset=["addr:street"])
```

```python
df_rent.drop_duplicates(subset=["name"], inplace=True)
```

```python
df_rent_prenaj = df_rent.loc[df_rent["sellCategory"]== "Prenájom"]
```

```python
df_rent_prenaj.drop(columns=["address", "price"], inplace=True)
```

```python
df_byty = df_byty[["addr:street","addr:housenumber","x","y","parcela","pocet_vla_spolu","ulica","cislo","ulica_cislo","Typ budovy","Mestská časť","MČ"]].drop_duplicates(subset=["addr:street"]).sample(frac = 1)[0:df_rent_prenaj.shape[0]].reset_index(drop=True)
```

```python
df_rent_prenaj.reset_index(drop=True, inplace=True)
```

```python
df_merge = df_rent_prenaj.join(df_byty)
```

```python
df_merge.to_csv("nehnutelnosti_cold_start.csv")
df_merge
```

```python
# COUNT ROOM COUNT
def compute_grade(df):
    if df['type'] == "1 izbový byt":
        return 1
    elif df['type'] == "2 izbový byt":
        return 2
    elif df['type'] == "3 izbový byt":
        return 3
    elif df['type'] == "4 izbový byt":
        return 4
    elif df['type'] == "5 a viac izbový byt":
        return 5
    elif df['type'] == "Iný byt":
        return 3
    elif df['type'] == "Loft":
        return 1
    elif df['type'] == "Garsónka":
        return 1
    elif df['type'] == "Apartmán":
        return 4
    elif df['type'] == "Mezonet":
        return 6
```

```python
df_merge['room_count'] = df_merge.apply(compute_grade, axis = 1)
```

```python
# filter valid items
df_merge = df_merge.loc[df_merge.singlePrice.str.contains("€/m²/mes.")]

```

```python
def pps_split(df):
    return float(df["singlePrice"].split("€/m²/mes.")[0].replace(",", "."))
```

```python
df_merge['price_per_meter'] = df_merge.apply(pps_split, axis = 1)
```

```python
df_merge['pople_couunt'] = round(df_merge["size"] / 15)
```

```python
df_merge['price_per_person'] = df_merge['price_per_meter'] * df_merge['size'] / df_merge["pople_couunt"]
```

```python
def image_path_split(df):
    return df["imagePath"].split("./images/")[1]
```

```python
df_merge["imagePath"] = df_merge.apply(image_path_split, axis = 1)
```

```python
df_merge.loc[df_merge["Typ budovy"] == "Iná budova", "Typ budovy"] = "Bytový dom"
df_merge.loc[df_merge["Typ budovy"] == "Budova obchodu a služieb", "Typ budovy"] = "Bytový dom"
df_merge.loc[df_merge["Typ budovy"] == "Iná budova", "Typ budovy"] = "Bytový dom"
df_merge.loc[df_merge["Typ budovy"] == "Polyfunkčná budova", "Typ budovy"] = "Bytový dom"
```

```python
df_merge.to_csv("nehnutelnosti_cold_start.csv")
```

```python
# INIT MAP
import folium
from folium.plugins import BeautifyIcon
from folium import plugins
m = folium.Map(location=[48.732681549999995, 21.244332875459712],zoom_start=13)
```

```python
# Add marker for Boulder, CO
# circle marker

for idx, byt in df_merge.iterrows():
    color = "black"

    icon_circle = BeautifyIcon(
        icon_shape='circle-dot',
        border_color=color,
        border_width=3,
    )
    folium.Marker(
        location=[byt["y"], byt["x"]], # coordinates for the marker (Earth Lab at CU Boulder)
        popup=byt["name"], # pop-up label for the marker
        # tooltip='circle',
        icon=icon_circle
    ).add_to(m)
```
